```python
ds_train[['RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']]=ds_train[['RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']].fillna(ds_train[['RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']].mean())
ds_test[['RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']]=ds_test[['RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']].fillna(ds_test[['RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']].mean())
```

```python
ds_train[['VIP','CryoSleep','Cabin']]=ds_train[['VIP','CryoSleep','Cabin']].fillna(value=0)
ds_train['Age']=ds_train['Age'].fillna(ds_train['Age'].mean())
ds_train['Destination']=ds_train['Destination'].fillna(method='bfill')
ds_train['HomePlanet']=ds_train['HomePlanet'].fillna(method='bfill')
ds_test[['VIP','CryoSleep','Cabin']]=ds_train[['VIP','CryoSleep','Cabin']].fillna(value=0)
ds_test['Age']=ds_train['Age'].fillna(ds_train['Age'].mean())
ds_test['Destination']=ds_train['Destination'].fillna(method='bfill')
ds_test['HomePlanet']=ds_train['HomePlanet'].fillna(method='bfill')
```

```python
ds_train=ds_train.drop(columns='Name',axis=1)
ds_test=ds_test.drop(columns='Name',axis=1)
```
